```python
class General_neuron: # שבה הסברתי מה אני רוצה שיהיה בתוך הדאטה טייפ (class) יצרתי דאטה טייפ חדש ויצרתי מחלקה 
    def __init__(self, firing_rate=0): #תכונה קצב הירי (במצב מנוחה)
        self.firing_rate=firing_rate
    def print_firing_rate (self): 
      print (self.firing_rate)
    def activate (self,stimulus): #פונקציה חדשה ,פונקציונליות - אקטיבציה 
        threshold=50 #(פוטנציאל פעולה) סף גירוי בנוירון שיגרום לאקטיבציה 
        self.stimulus=stimulus
        if stimulus > threshold:
            self.firing_rate="high" #קצב הירי גבוה אם הגירוי גבוה מהסף 
        else: 
            self.firing_rate="low" #קצב הירי נמוך אם גירוי נמוך מהסף 
        return self.firing_rate 
    pass
```

```python
class Sensory_Neuron(General_neuron): # יוצרת תתאי סוגים של נוירונים - נוירון סנסורי, אבייקט חדש 
    def __init__(self,firing_rate, activate): #נוירון סנסורי שומר את הכונות של נוירון הראשי
        super().__init__(firing_rate, activate) # יהיו גם בתת סוג, יורש את התכונות של המחלקהGeneral_Neuron כל מה שהיה במחלקה הראשית 
        self.id=id #אפשר לעשות שינויים
    def __init__(self, receptor_type,stimulus_type): #הגדרת פונקציה חדשה באובייקט נוירון סנסורי 
        self.receptor_type=receptor_type
        self.stimulus_type=stimulus_type
        if self.receptor_type=="photoreceptor": #אם הנוירון שלי מסוג פוטורצפטור 
         self.stimulus_type="light" #אז הגירוי עליו יגיב הוא אור 
        else:
         self.receptor_type=="mechanoreceptors" #אם הנוירון שלי מסוג מכנורצפטור
        self.stimulus_type="pressure" #אז יגיב לגירוי לחץ 
    def Sense_Stimulus(self, stimulus_strength): # מגדירה פונקציה חדשה - רגישות של גירוי  
       self.stimulus_strength=stimulus_strength 
       self.firing_rate=stimulus_strength #רגישות הגירוי תלויה בקצב הירי (פונקציה שכבר הגדרתי במחלקה הראשית שלי)
       print (self.firing_rate)
pass
```

```python
class Motor_Neuron(General_neuron): #אובייקט חדש ,(General_neuron) נוירון מוטורי שהוא תת סוג של נוירון הראשי 
   def __init__(self,firing_rate, activate,target_muscule): 
       super().__init__(firing_rate, activate) #נוירון מוטורי שומר את התכונות של הנוירון הראשית, יורש את התכונות של המחלקה
       self.id=id #אפשר לעשות שינויים 
       self.target_muscule=target_muscule #אטריביוט שריר המטרה
   def print_target_muscule (self): 
      print (f"target muscule: {self.target_muscule}")
      motor_neuron1=Motor_Neuron(target_muscule="skeletal_muscle") 
      motor_neuron1.print_target_muscule()
      motor_neuron2=Motor_Neuron(target_muscule="muscle_spindle")
      motor_neuron2.print_target_muscule()
   def Control_Muscle (self,activation_level): #יצירת פונקציה של קונטרול שריר על בסיס רמת האקטיבציה 
      self.activation_level=activation_level
      if activation_level>0:
       print(f"The {self.target_muscule} is activated with level of {activation_level}.")
      else:
       print(f"The {self.target_muscule} is not activated with level of {activation_level}.")
       pass 
```

```python
class Photoreceptor(Sensory_Neuron): #יצירת מחלקה של פוטורצפטור שזה סוג של נוירון סנסורי 
   def __init__(self,firing_rate, receptor_type="light"):
      super().__init__(firing_rate, receptor_type, "light") #ירושה של תכונות של  נוירון סנסורי 
   def light_detection (self, light_intensity):
      self.light_intensity=light_intensity
      if light_intensity > 50: #אם עוצמת האור גדולה 
         self.firing_rate = "high" #קצב הירי גבוה 
      elif 20 < light_intensity <= 50: #אם עוצמת האור בין 50 ל20
         self.firing_rate = "medium" #קצב הירי בינוני
      else:
         self.firing_rate = "low" #אם עוצמת האור קטנה, אז קצב הירי נמוך
         print(f"Light intensity: {light_intensity}, Firing rate: {self.firing_rate}") 
pass 
```

```python
class Mechanoreceptor(Sensory_Neuron): #יצירת מחלקה מכנורספטור שזה סוג של נוירון סנסורי 
   def __init__(self, firing_rate, stimulus_type="pressure"): #סוג של גירוי הוא חלץ
      super().__init__(firing_rate, stimulus_type, "pressure") #יורש תכונות של נוירון סנסורי
   def Pressure_Detection (self, pressure_strength): #קצב הירי לפי עוצמת הלחץ
      self.pressure_strength=pressure_strength
      if pressure_strength>50:
         self.firing_rate="high" #קצב הירי גבוה אם עוצמת הלחץ גבוה מ50
      elif 20 < pressure_strength <= 50:
         self.firing_rate="medium" # קצב הירי בינוני אם עוצמת הלחץ בין 50 ל20 
      else:
          self.firing_rate="low" #קצב הירי נמוך אם עוצמת הלחץ קטן מ50 
          print(f"Pressure stength:{pressure_strength}, Firing rate: {self.firing_rate}")
pass 
```


# Q1

```python
f=5 #תדירות
sample_rate=100 #העברתי למספר שנוח לחישוב מבחית הדגימה, דגמתי 100 נקודות ,לכל 0.01 שניה
duration=1 #הסיגנל פעיל במשך שניה אחת 
t = np.arange(0, duration, 1/sample_rate) #עם דגימה בשניות (duration) וקטור בפרק זמן בין 0 ל1 
signal=np.sin(2*np.pi*f*t) #יצירת סיגנל סינוסי 
min(signal) #בדיקה מהיא נקודת מינימום 
max(signal) #בדיקה מהיא נקודה מקסימום 
downsampled_signal = signal[::20] #דגימה קטנה יותר ,דגימה של 5 פריטים מתוך הסיגנל הכללי כל 20 דגימות - 5 נקודות מתוך 100
downsampled_t=t[::20]# דגימה של 5 נקודות ב t - כל 20 דגימות - 5 נקודות מתוך 100 
downsampled_min = [min] if min in downsampled_signal else [downsampled_signal[np.argmin(downsampled_signal)]] #בדיקה האם ערך מינימום נמצא בדגימה 
downsampled_max = [max] if max in downsampled_signal else [downsampled_signal[np.argmax(downsampled_signal)]] #האם ערך מקסימום נמצא בדגימה 
print("Length of downsampled signal:", len(downsampled_signal)) #כדי לראות את התוצאה מהליסט שלי 
print ("downsamples min:",downsampled_min)
print ("downsamples max:",downsampled_max)
```

```python
plt.plot(signal) #matplotlib יצירת גרף באמצעות ספריה 
```

```python
plt.plot(t, signal, label='signal') #יצירת גרף 
plt.scatter(downsampled_t, downsampled_signal, color='orange', label='5 points of sample') #דגימת נקודות פיזור על הגרף - ציון עבור איזה נקודות ספציפית, צבע ושם בגרף 
plt.legend(), plt.grid(), plt.show() #הדגשת 5 נקודות שדגמתי, יצירת מקרה על הגרף (איפה שמופיעים הליבלים שלי)
```
